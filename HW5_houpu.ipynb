```python
import numpy as np
from scipy import stats

n1 = 15
mean1 = 2.8
std1 = 1.6

n2 = 19
mean2 = 2.4
std2 = 1.7

df1 = n1 - 1
df2 = n2 - 1
df_total = n1 + n2 - 2

pooled_var = (df1 * (std1**2) + df2 * (std2**2)) / df_total
se_diff = np.sqrt(pooled_var * (1/n1 + 1/n2))

t_stat = (mean1 - mean2) / se_diff

p_val_one_tailed = stats.t.sf(t_stat, df_total)


print(f"Pooled Variance: {pooled_var:.4f}")
print(f"p-value: {p_val_one_tailed:.4f}")

# 7. 
alpha = 0.05
if p_val_one_tailed < alpha:
    print(f"conclusion: value < {alpha}Reject the HO hypothesis. There is striking evidence that high-income families have fewer children")
else:
    print(f"conclusion: value >= {alpha}we cannot reject HO hypothesis")
```

```python
import pandas as pd

no_bin = [4, 4, 4, 4, 4, 4, 4, 5, 8, 9, 9, 9, 9, 12, 12, 13, 14, 14, 14, 14, 15, 23]
with_bin = [4, 5, 8, 8, 8, 9, 9, 9, 12, 14, 14, 15, 16, 19, 23, 28, 40, 43, 129, 130]

combined = np.concatenate([no_bin, with_bin])
labels = ['No Bin'] * len(no_bin) + ['With Bin'] * len(with_bin)
df_ranks = pd.DataFrame({'Value': combined, 'Group': labels})
df_ranks['Rank'] = df_ranks['Value'].rank(method='average') 
df_ranks
```

e)

```python
rank_sum_no = df_ranks[df_ranks['Group'] == 'No Bin']['Rank'].sum()
rank_sum_with = df_ranks[df_ranks['Group'] == 'With Bin']['Rank'].sum()
n1, n2 = len(no_bin), len(with_bin)

print(f"e) sample size: No Bin(n1)={n1}, With Bin(n2)={n2}")
print(f"e) rank sum: No Bin={rank_sum_no}, With Bin={rank_sum_with}")
```

f)

```python
u1 = n1 * n2 + (n1 * (n1 + 1)) / 2 - rank_sum_no
print(f"f) Mann-Whitney Ut (No Bin): {u1}")
```

g)

```python
u2 = n1 * n2 + (n2 * (n2 + 1)) / 2 - rank_sum_with
print(f"g) Mann-Whitney Ut (With Bin): {u2}")
```

h)

```python
u_stat = min(u1, u2)
print(f"h) Value of test statistic U: {u_stat}")
```

```python
#  P  -  i
res = stats.mannwhitneyu(no_bin, with_bin, alternative='two-sided')
print(f"i) P-value: {res.pvalue:.4f}")

# 
if res.pvalue < 0.05:
    print("conclusion: P < 0.05There is a significant difference in the distribution of paper use between the two groups")
else:
    print("conclusion: P >= 0.05There is a no significant difference in the distribution of paper use between the two groups")
```

```python
import numpy as np
from scipy import stats
import matplotlib.pyplot as plt

females = np.array([0.291, 0.208, 0.241, 0.437, 0.228, 0.256, 0.208, 0.234, 0.320, 0.340, 0.150])
males = np.array([0.185, 0.222, 0.149, 0.187, 0.191, 0.219, 0.132, 0.144, 0.140])

log_females = np.log(females)
log_males = np.log(males)

mean_log_f = np.mean(log_females)
mean_log_m = np.mean(log_males)
std_log_f = np.std(log_females, ddof=1)
std_log_m = np.std(log_males, ddof=1)

t_stat, p_val = stats.ttest_ind(log_females, log_males, equal_var=True)

print(f"female mosquitoes (n=11): mean log = {mean_log_f:.4f}, sd = {std_log_f:.4f}")
print(f"male mosquitoes (n=9):  mean log = {mean_log_m:.4f}, sd = {std_log_m:.4f}")
print(f"t : {t_stat:.4f}")
print(f"degree of freedom: {len(females) + len(males) - 2}")
print(f"p-value: {p_val:.6f}")

alpha = 0.05
if p_val < alpha:
    print(f"Conclusion: P < {alpha}There is a significant difference in the average body weight of female and male mosquitoes.")
else:
    print(f"Conclusion: P >= {alpha}There is no significant difference in the average body weight of female and male mosquitoes.")
```
