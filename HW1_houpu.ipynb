```python
import numpy as np
import pandas as pd
from scipy import stats

stream_lengths = np.array([
    100, 426, 322, 466, 112, 155, 388, 1155,
    234, 324, 556, 221, 18, 133, 177, 441
])
```

```python
mean_val = stream_lengths.mean()
mean_val
```

```python
std_val = stream_lengths.std(ddof=1)
std_val
```

```python
kurtosis_val = stats.kurtosis(stream_lengths)
kurtosis_val
```

```python
skew_val = stats.skew(stream_lengths)
skew_val
```

The dataset shows a positively skewed distribution, as indicated by the skewness value of 1.8252570260382082. This suggests that most stream link lengths are relatively short.

Overall, the combination of positive skewness and high kurtosis suggests that the stream length data are not normally distributed. Instead, they exhibit a right-skewed, heavy-tailed pattern that is characteristic of river networks.

# Question 3

```python
import numpy as np
import pandas as pd

data = {
    "rainfall_midpoint": [15, 25, 35, 45, 55, 65],   # given / assumed midpoints
    "number": [5, 10, 12, 11, 3, 2]
}

df = pd.DataFrame(data)

N = df["number"].sum()

# calculate grouped mean
grouped_mean = (df["rainfall_midpoint"] * df["number"]).sum() / N

# calculate grouped variance
grouped_variance = (df["number"] * (df["rainfall_midpoint"] - grouped_mean) ** 2).sum() / N

grouped_mean, grouped_variance
```

According the result above, The group mean is <font color='red'>35.69767441860465</font>, and the group variance is <font color='red'>164.62952947539208</font>

```python
import numpy as np

# commuting times
com_t = np.array([23, 43, 42, 7, 23, 11, 23, 55])

# mean
com_t_mean = com_t.mean()

# sample standard deviation (ddof=1)
com_t_std = com_t.std(ddof=1)

# coefficient of variation
coefficient_var = com_t_std / com_t_mean

coefficient_var
```

# Question 5

```python
import numpy as np
import pandas as pd

df = pd.DataFrame({
    "x": [3, 2, 1, 6, 2],
    "y": [7, 2, 1, 2, 5],
    "pop": [40, 10, 50, 30, 20]
})

df
```

a)

```python
mean_x = df["x"].mean()
mean_y = df["y"].mean()

distance_mean = np.sqrt((df["x"]-mean_x)**2 + (df["y"]-mean_y)**2)
aggregate_distance_meancenter = (distance_mean * df["pop"]).sum()
aggregate_distance_meancenter
```

According the result above, The aggregate travel distance far from mean center is <font color='red'>450.9091493345912</font>

b)

```python
median_x = df["x"].median()
median_y = df["y"].median()

distance_median = np.sqrt((df["x"]-median_x)**2 + (df["y"]-median_y)**2)
aggregate_distance_mediancenter = (distance_median * df["pop"]).sum()
aggregate_distance_mediancenter
```

According the result above, The aggregate travel distance far from median center is <font color='red'>454.67145866236615</font> Compare to the result a), the aggregate travel distance far from median center is higher.

c)

```python
tot_pop = df["pop"].sum()
standard_distance = np.sqrt(
    (
        df["pop"] * (
            (df["x"] - mean_x)**2 + (df["y"] - mean_y)**2
        )
    ).sum() / tot_pop
)

standard_distance
```

According the result above, The standard distance is <font color='red'>3.0832882879592476</font>.

# Question 6

```python
import numpy as np
import pandas as pd

df = pd.DataFrame({
    "city": ["A", "B", "C", "D", "E"],
    "x": [3.3, 1.1, 5.5, 3.7, 1.1],
    "y": [4.3, 3.4, 1.2, 2.4, 1.1],
    "pop": [34000, 6500, 8000, 5000, 1500]
})

df
```

a)

```python
W = df["pop"].sum()

mean_x_w = (df["x"] * df["pop"]).sum() / W
mean_y_w = (df["y"] * df["pop"]).sum() / W

mean_x_w, mean_y_w
```

```python
mean_x = df["x"].mean()
mean_y = df["y"].mean()

mean_x, mean_y
```

According the result above, the <font color='red'>unweight mean center</font> locate at the left bottom of <font color='red'>weighted mean center</font>.

b)

```python
df["dist_to_weighted_mean_center"] = np.sqrt(
    (df["x"] - mean_x_w)**2 +
    (df["y"] - mean_y_w)**2
)

df[["city", "dist_to_weighted_mean_center"]]
```

c)

```python
weighted_standard_distance = np.sqrt(
    (df["pop"] *((df["x"] - mean_x_w)**2 +(df["y"] - mean_y_w)**2)
    ).sum() / W
)

weighted_standard_distance
```

According the result above, the weighted standard distance center is <font color='red'>1.6894639474555428</font>.

```python
import numpy as np
import matplotlib.pyplot as plt
from collections import Counter

salaries = np.array([
    27, 60, 143, 25, 24, 27, 42, 98, 29, 30, 97, 59, 26, 43, 27,
    44, 25, 27, 26, 61, 145, 26, 40, 27, 46, 28, 24
])
```

a)

```python
mean_val = salaries.mean()
median_val = np.median(salaries)
mode_val = Counter(salaries).most_common(1)[0][0]
range_val = salaries.max() - salaries.min()
std_val = salaries.std(ddof=1)

mean_val, median_val, mode_val, range_val, std_val
```

According the result above, the mean is <font color='red'>47.25925925925926</font>, the median is <font color='red'>29.0</font>, the mode is <font color='red'>27</font>, the range is <font color='red'>121</font>,and the standara deviation <font color='red'>34.44682925298126</font>.

b)



```python
z_scores_first_five = (salaries[:5] - mean_val) / std_val
z_scores_first_five
```

c)

```python
# Absolute frequency histogram

plt.figure()
plt.hist(salaries, bins=10)
plt.xlabel("Salary ($/hour)")
plt.ylabel("Frequency")
plt.title("Histogram of Salaries (Absolute Frequency)")
plt.show()
```

```python
# Relative frequency histogram
plt.figure()
plt.hist(salaries, bins=10, density=True)
plt.xlabel("Salary ($/hour)")
plt.ylabel("Relative Frequency")
plt.title("Histogram of Salaries (Relative Frequency)")
plt.show()
```
